```python
def strong_connect(Graph):
    """
    This function will test if a Digraph is strongly connected.
    If the graph is strongly connected it will display a message saying it is.
    If the graph is not strongly connected it will pass the graph to the "weakly_connected"
    function for testing
    param: Graph (a graph object)
    return: None
    """
    if Graph.is_strongly_connected() == True:
        print"This graph is strongly connected."
    else:
        print"This graph is not strongly connected."
        weakly_connected(Graph)
```

```python
def weakly_connected(Graph):
    """
    This function is going to test if a DiGraph is weakly connected.
    It will do this be turning the DiGraph into a undirected graph.
    It will than test connectivaty of the indirected graph.
    The function will print to screen the results
    Param: Graph (is a graph object)
    Return: None
    """
    A = Graph.to_undirected()
    if A.is_connected() == True:
        print"The graph is weakly connected"
    else:
        print"The graph is not weakly connected"
```

```python
weakly_connected(E)
```

```python
def strong_connect_component(Graph):
    """
    This function will print to screen a list of strongly connected component of a graph
    param: Graph (is a graph object)
    return: None
    """
    print"A list of the Graphs strongly connected component {}.".format(Graph.strongly_connected_components())
    
```

```python
def edge_vertex_cut(Graph):
    """
    Will print to screen a list of the edge cut and vertex cut for a graph
    param: Graph (is a graph object)
    return: None
    """
    cut_edges_list = []
    A1 = Graph.edges(labels=False)

    for i in A1:
        if Graph.is_cut_edge(i) == True:
            cut_edges_list.append(i)

    if len(cut_edges_list) == 0:
        print"There are no cut edges in this graph"
    else:
        print"This a list of cut edges in the graph {}.".format(cut_edges_list)

    cut_vertices_list = []
    B = Graph.vertices()
    for i in B:
        if Graph.is_cut_vertex(i) == True:
            cut_vertices_list.append(i)
    
    if len(cut_vertices_list) == 0:
        print"There is no cut vertex is this graph"
    else:
        print"This a list of cut vertices in the graph {}.".format(cut_vertices_list)
    
```

```python
edge_vertex_cut(E)
```
